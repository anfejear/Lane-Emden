```python
from sympy import *
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def coeficientes(a,k,n):
    sumatoria = 0
    for i in np.arange(1,k+1):
        sumatoria = sumatoria + (i*n+i-k)*(k-i+1)*(3+2*k-2*i)*a[i]*a[k-i+1]
    return np.append(a,sumatoria/(k*(k+1)*(2*k+3)))
```

```python
def solucion(n,p,x):
    sumatoria = 0
    a = np.array([0,-1/6])
    for k in np.arange(0,p):
        sumatoria = sumatoria + a[k]*(x**(2*k))
        if(k >= 1):
            a = coeficientes(a,k,n)
    return 1 + sumatoria
```

```python
x = np.arange(0,3,0.01)
fig =  plt.figure(figsize=(10,8))
plt.plot(x,1-x**2/6,'o',label='Sol. Analítica')
plt.plot(x,solucion(0,20,x),'r',label='Sol. Numérica')
plt.xlabel('x')
plt.ylabel('y')
plt.title('n=0')
plt.legend()
plt.grid()
plt.show()
fig.savefig('n0.svg')
```
